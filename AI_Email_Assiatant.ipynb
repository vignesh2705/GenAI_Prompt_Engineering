```python
from groq import Groq

# Replace with your Groq API Key
client = Groq(api_key="YOUR_GROQ_API_KEY")
```

```python
# -------------------------------------------------
# Prompt 1: Basic Prompt
# -------------------------------------------------
basic_prompt = """
Generate a professional interview confirmation email.

Requirements:
- Accept the interview invitation.
- Confirm attendance on 20 July 2026 at 10:00 AM.
- Include a thank-you note.
"""

response1 = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "user", "content": basic_prompt}
    ],
    temperature=0.5
)

print("=" * 70)
print("INTERVIEW CONFIRMATION EMAIL (Basic Prompt)")
print("=" * 70)
print(response1.choices[0].message.content)


# -------------------------------------------------
# Prompt 2: Improved Formal Prompt
# -------------------------------------------------
formal_prompt = """
Act as a professional job applicant.

Write a formal interview confirmation email to the HR team.

Requirements:
1. Formally accept the interview invitation.
2. Confirm attendance on 20 July 2026 at 10:00 AM.
3. Express sincere gratitude for the opportunity.
4. Maintain a respectful, professional, and formal tone.
5. Include:
   - Subject line
   - Greeting
   - Email body
   - Professional closing with the applicant's name.
6. Keep the email concise (150–200 words).
7. Return only the email without any explanation.
"""

response2 = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "user", "content": formal_prompt}
    ],
    temperature=0.3
)

print("\n" + "=" * 70)
print("INTERVIEW CONFIRMATION EMAIL (Improved Formal Prompt)")
print("=" * 70)
print(response2.choices[0].message.content)
```
