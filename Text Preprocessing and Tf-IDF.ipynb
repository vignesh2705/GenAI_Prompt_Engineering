## Product Review Text Preprocessing and TF-IDF Feature Extraction

```python
# Sample Dataset
reviews = [
    "This laptop is lightweight and performs very well.",
    "Battery life is poor, but the display is amazing.",
    "Excellent performance with great battery backup.",
    "I did not like the keyboard quality."
]

# Display original reviews
print("Original Reviews:")
for i, review in enumerate(reviews):
    print(f"{i+1}. {review}")
```

### Step 1: Convert all reviews to lowercase and remove punctuation.

```python
import string

cleaned_reviews = []
for review in reviews:
    # 1. Convert to lowercase
    review = review.lower()
    # 2. Remove punctuation
    review = review.translate(str.maketrans('', '', string.punctuation))
    cleaned_reviews.append(review)

print("Reviews after lowercasing and punctuation removal:")
for i, review in enumerate(cleaned_reviews):
    print(f"{i+1}. {review}")
```

### Step 2: Tokenize the reviews and remove English stop words.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary NLTK data (if not already downloaded)
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

# Explicitly download 'punkt_tab' as suggested by the error message
# This resource seems to be specifically required by word_tokenize in this context
try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')

stop_words = set(stopwords.words('english'))

final_cleaned_reviews = []
for review in cleaned_reviews:
    # 3. Tokenize the reviews
    tokens = word_tokenize(review)
    # 4. Remove English stop words
    filtered_tokens = [word for word in tokens if word not in stop_words]
    final_cleaned_reviews.append(' '.join(filtered_tokens))

print("Cleaned Reviews (tokenized and stop words removed):")
for i, review in enumerate(final_cleaned_reviews):
    print(f"{i+1}. {review}")
```

### Step 3: Apply TF-IDF Vectorization and display the results.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 6. Apply TF-IDF Vectorization
# Use a custom tokenizer and disable internal preprocessing since the text is already cleaned.
vectorizer = TfidfVectorizer(
    lowercase=False,  # Text is already lowercased
    preprocessor=lambda x: x, # No additional preprocessing needed
    tokenizer=lambda x: x.split(), # Use space to split already tokenized words
    stop_words=None, # Stop words already removed by NLTK
    token_pattern=None # Disable default regex pattern since we are providing a tokenizer
)
tfidf_matrix = vectorizer.fit_transform(final_cleaned_reviews)

# 7. Display results
# Cleaned Reviews
print("\n--- Cleaned Reviews ---")
for i, review in enumerate(final_cleaned_reviews):
    print(f"{i+1}. {review}")

# TF-IDF Feature Names
feature_names = vectorizer.get_feature_names_out()
print("\n--- TF-IDF Feature Names ---")
print(feature_names)

# TF-IDF Matrix
print("\n--- TF-IDF Matrix (Sparse) ---")
print(tfidf_matrix)

# Optionally, display the TF-IDF matrix as a DataFrame for better readability
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)
print("\n--- TF-IDF Matrix (DataFrame) ---")
display(tfidf_df)
```
